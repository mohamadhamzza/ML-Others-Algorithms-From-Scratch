```python
def bubble_sort(persons,x="name"):
    swap=False
    while not swap:
        swap=True
        for i in range(1,len(persons)):
            if persons[i-1][x]>persons[i][x]:    
                swap=False        
                persons[i],persons[i-1]= persons[i-1],persons[i]
    return pd.DataFrame(persons)
              
bubble_sort(persons,x="name")
```

```python
def selection_sort ( persons,x="name"):
    
    for i in range (len(persons)):
        mn=i
        for j in range(i+1, len(persons)):
            if persons[j][x]<persons[mn][x]:  
                mn=j
                persons[i],persons[mn]=persons[mn],persons[i]
    return pd.DataFrame(persons)
            
selection_sort(persons,x="name")
```

```python
def merge_sort(persons,feature_name):
    if len (persons)<=1:
        return persons
    mid=len(persons)//2
    left=persons[:mid]
    right=persons[mid:]
    
    left=merge_sort(left,feature_name)
    right=merge_sort(right,feature_name)
    
    return merge_two_list(left,right,feature_name)

def merge_two_list(a,b,feature_name):
    srd_lst=[]
    len_a=len(a)
    len_b=len(b)
    i=j=0
    while i < len_a and j < len_b:
        if a[i][feature_name]<= b[j][feature_name]:
            srd_lst.append(a[i])
            i+=1
        else:
            srd_lst.append(b[j])
            j+=1
    while i < len_a:
        srd_lst.append(a[i])
        i+=1
        
    while j < len_b:
        srd_lst.append(b[j])
        j+=1
    return srd_lst  

#d={"A":6,"B":2,"c":3,"S":1,"D":4,"E":5}

#values = merg_sort(list(d.values()))


#d2=dict()

#for value in values :
#    for key in d:
        
#        if d[key]==value:
#            d2[key]=value
#print(d2)
   
pd.DataFrame(merge_sort(persons,feature_name="name"))
```

```python
def builtin_sort(persons, x="name"):
    return pd.DataFrame(sorted(persons, key=lambda k: k[x]))
```

```python
import time

# Generate the fake profiles
fake = Faker()
Faker.seed(4321)
persons = [fake.profile() for _ in range(100)]  # Increased the size for a better comparison

# Define a function to measure the execution time
def measure_time(sort_func, persons, x):
    start = time.time()
    sort_func(persons.copy(), x)
    end = time.time()
    return end - start

# Measure and print the execution times
bubble_sort_time = measure_time(bubble_sort, persons, "name")
selection_sort_time = measure_time(selection_sort, persons, "name")
merge_sort_time = measure_time(merge_sort, persons, "name")
builtin_sort_time = measure_time(builtin_sort, persons, "name")

print(f"Bubble Sort Time: {bubble_sort_time:.6f} seconds")
print(f"Selection Sort Time: {selection_sort_time:.6f} seconds")
print(f"Merge Sort Time: {merge_sort_time:.6f} seconds")
print(f"Built-in Sort Time: {builtin_sort_time:.6f} seconds")
```

```python

# Define a function to measure execution time
def measure_time(sort_func, persons, x):
    start = time.time()
    sort_func(persons.copy(), x)
    end = time.time()
    return end - start

# Generate datasets of varying sizes
fake = Faker()
Faker.seed(4321)
sizes = [10, 50, 100, 500, 1000]
datasets = {size: [fake.profile() for _ in range(size)] for size in sizes}

# Run experiments and collect results
results = {size: {"Bubble Sort": [], "Selection Sort": [], "Merge Sort": [], "Built-in Sort": []} for size in sizes}

for size in sizes:
    for _ in range(10):  # Run each test 10 times
        persons = datasets[size]
        results[size]["Bubble Sort"].append(measure_time(bubble_sort, persons, "name"))
        results[size]["Selection Sort"].append(measure_time(selection_sort, persons, "name"))
        results[size]["Merge Sort"].append(measure_time(merge_sort, persons, "name"))
        results[size]["Built-in Sort"].append(measure_time(builtin_sort, persons, "name"))

avg_times = {}
for size, result in results.items():
    avg_times[size] = {}
    for algo, times in result.items():
        avg_times[size][algo] = sum(times) / len(times)


print("Average Execution Times (in seconds):")
df = pd.DataFrame(avg_times).T
print(df)

plt.figure(figsize=(10, 6))
for algo in ["Bubble Sort", "Selection Sort", "Merge Sort", "Built-in Sort"]:
    plt.plot(sizes, [avg_times[size][algo] for size in sizes], label=algo)
plt.xlabel('Input Size')
plt.ylabel('Average Execution Time (seconds)')
plt.title('Comparison of Sorting Algorithms')
plt.legend()
plt.grid(True)
plt.show()
```
